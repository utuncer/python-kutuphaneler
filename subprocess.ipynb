```python
import os
import subprocess as sp

def open_app(path:str):
    if os.path.exists(path):
        sp.Popen(path, stdout=sp.DEVNULL, stderr=sp.DEVNULL) # Belirtilen yoldaki uygulamayÄ± baÅŸlatÄ±r; Ã§Ä±ktÄ±larÄ± ve hatalarÄ± yoksayarak (DEVNULL) konsolun kirlenmesini Ã¶nler.
    else:
        print(f"Error: Application not found at \"{path}\"")

# ~ kullanÄ±cÄ± dizinini temsil eder.
# r string baÅŸÄ±nda kullanÄ±ldÄ±ÄŸÄ±nda, kaÃ§Ä±ÅŸ karakterlerini yok sayar.

notion_path = os.path.expanduser(r"~\AppData\Local\Programs\Notion\Notion.exe") # Notion uygulamasÄ±nÄ±n tam dosya yolunu tanÄ±mlar.

anki_path = os.path.expanduser(r"~\AppData\Local\Programs\Anki\anki.exe") # Anki uygulamasÄ±nÄ±n tam dosya yolunu tanÄ±mlar.

notepad_path = r"C:\Windows\system32\notepad.exe" # Notepad uygulamasÄ±nÄ±n sabit dosya yolunu tanÄ±mlar.

while True:
    print("*"*30)
    choice = input("Make a choice: \n1-Notepad\n2-Notion\n3-Anki\n-> ")
    if choice == "1":
        open_app(notepad_path)
    elif choice == "2":
        open_app(notion_path) 
    elif choice == "3":
        open_app(anki_path)
    else:
        print(f"{choice} is not a valid choice.")
    exit_choice = input("Do you want to exit? (y/n): ")
    if exit_choice.lower() == 'y':
        break
    elif exit_choice.lower() == 'n':
        continue
    else:
        print(f"{exit_choice} is not a valid choice. Continuing...")
        continue
```
