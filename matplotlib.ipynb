```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel(r'excel\teknolojik_urunler_zamanli.xlsx')
df
```

```python
df['Tarih'] = pd.to_datetime(df['Tarih'])
df.set_index('Tarih', inplace=True)
df
```

```python
df['Satış'].plot(title='Satışların zaman içerisindeki değişimi',xlabel='Tarih',ylabel='Satış')
```

```python
aylik_satis = df.resample('ME')['Satış'].sum()
aylik_satis.plot(kind='bar', title="Aylık toplam satış",xlabel="Tarih",ylabel='Satış')
```

```python
kategori_satis = df.groupby('Kategori')['Satış'].sum()
kategori_satis.plot(title='Kategorilere göre satış dağılımı',autopct='%1.1f%%',kind='pie',ylabel='')
```

```python
df.plot(kind='scatter',x = 'Fiyat (TL)', y='Satış',title="Fiyat ve Satış İlişkisi")
```

```python
df['Fiyat (TL)'].plot(kind="hist",bins=10,title="Fiyat dağılımı",xlabel="Fiyat (TL)")
```

```python
aylik_satis = df.resample('ME')['Satış'].sum()
aylik_satis.plot(kind='line',title='Aylık Satış Miktarları',xlabel='Aylar', ylabel='Satış Sayısı')
```

```python
# 1. ADIM: GERÇEK VERİYİ GÖRSELLEŞTİRME
# Mavi noktaları (scatter) çiziyoruz. Bunlar yaşanan gerçeği gösterir.
df.plot(kind='scatter', x='Fiyat (TL)', y='Satış', title='Fiyat ve Satış İlişkisi')

# 2. ADIM: MATEMATİKSEL MODELİ KURMA (Dedektiflik)
# np.polyfit: Verilere bakar ve "En az hatayla bu noktaların ortasından geçen doğrunun eğimi ne olmalı?" diye sorar.
# Çıktı (z): Bize bir sonuç listesi vermez. Bize formülün sayılarını verir. 
# Örn: z = [-0.002, 22] (Yani: Eğim -0.002, Başlangıç noktası 22)
z = np.polyfit(df['Fiyat (TL)'], df['Satış'], 1) 

# 3. ADIM: FORMÜLÜ MAKİNEYE DÖNÜŞTÜRME
# np.poly1d: Yukarıdaki 'z' katsayılarını alır ve çalışan bir fonksiyona çevirir.
# Artık 'p' değişkeni, içine fiyat atınca tahmini satış veren bir hesap makinesidir.
p = np.poly1d(z)

# 4. ADIM: İDEAL ÇİZGİYİ ÇİZME (Kırmızı Çizgi)
# Buradaki mantık şudur: 
# X ekseni: Yine aynı fiyatları kullan.
# Y ekseni: Fiyatları 'p' makinesine at ve ÇIKAN SONUÇLARI (tahminleri) kullan.
# Gerçek satışları değil, formülün ürettiği kusursuz sonuçları çizdirdiğimiz için ip gibi düz bir çizgi çıkar.
plt.plot(df['Fiyat (TL)'], p(df['Fiyat (TL)']), color='red')

# Grafiği ekrana bas.
plt.show()
```

```python
# --- 1. ADIM: ARALIKLARI VE İSİMLERİ TANIMLAMA ---

# bins: Sayı doğrusu üzerine atılan çentiklerdir.
# Aralıklar şöyle oluşur: 0-2000, 2000-5000, 5000-10000 vb.
bins = [0, 2000, 5000, 10000, 20000, 30000]

# labels: Oluşan boşluklara (aralıklara) verilecek isimler.
# KURAL: 'bins' listesindeki eleman sayısından her zaman 1 eksik olmalıdır.
labels = ['Düşük', 'Orta', 'Yüksek', 'Çok Yüksek', 'Lüks']


# --- 2. ADIM: KATEGORİZE ETME (pd.cut) ---

# df['Fiyat (TL)'] sütunundaki sayıları al, 'bins' sınırlarına göre kes,
# denk gelen aralığa 'labels' içindeki ismi yapıştır ve yeni bir sütuna kaydet.
df['Fiyat Kategorisi'] = pd.cut(df['Fiyat (TL)'], bins=bins, labels=labels)

# Tablonun son halini ekranda göster (Yeni sütunu kontrol etmek için)
display(df)


# --- 3. ADIM: ANALİZ VE GÖRSELLEŞTİRME ---

# Zincirleme işlem (Method Chaining):
# 1. .groupby: Veriyi 'Fiyat Kategorisi'ne göre gruplara ayır.
# 2. ['Satış']: Sadece satış adetlerini dikkate al.
# 3. .sum(): Her grubun satışlarını topla.
# 4. .plot(): Çıkan sonucu grafiğe dök.
df.groupby('Fiyat Kategorisi')['Satış'].sum().plot(
    kind='bar',                                     # Grafik türü: Sütun (Bar)
    title='Fiyat Kategorisine Göre Toplam Satışlar', # Grafiğin üst başlığı
    xlabel='Fiyat Kategorisi',                      # Yatay eksen etiketi
    ylabel='Toplam Satış'                           # Dikey eksen etiketi
)
```
